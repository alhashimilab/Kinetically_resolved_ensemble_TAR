```python
# Dependency Function Definitions:
def calc_rmsd(x, y):
    if len(x) == 0:
        return 0.0
    chi2 = sum((y - x) ** 2)
    rmsd = np.sqrt(chi2 / len(y))
    return rmsd

def calc_r2(x, y):
    if len(x) == 0:
        return 0.0
    correlation = r2_score(y, x)
    return correlation

def diff(list1, list2):
    c = set(list1).union(set(list2))  # or c = set(list1) | set(list2)
    d = set(list1).intersection(set(list2))  # or d = set(list1) & set(list2)
    common = []
    for element in list1:
        if element in list2:
            common.append(element)
    return list(c - d), common
```

```python
# FARFAR-Unbiased wt-TAR RDC Optimized Ensemble (N=20) (Lowest RMSD Orca5 with VTSCF)
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to ensemble

dfp = pd.read_csv("data/Orca_DFT/wtTAR_GS_ES1G34syn_Unbiased_LowestRMSD_DFT_VTSCF_OutlierFiltered.csv").drop_duplicates()
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])

# Indices of the conformers (relative position in lib)
indices = np.array([332,  367,  636,  683,  686,  997, 1196, 2295, 2615, 2751, 3051, 
                    4151, 4348, 4460, 4793, 4801, 6962, 8322, 9002, 9615], dtype=int)
ES_indices = (indices >= 5000)
GS_indices = (indices < 5000)

# List of resonances & residues to plot for this analysis 
resonances = ["C1'", "C4'", "C5'", "C8", "C6"]
residue_nos = [29, 30, 31, 32, 33, 34, 35, 36]
residue_names = ["C", "C", "U", "G", "G", "G", "A", "G"]

DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

for resonance in resonances:
    dfp_f = dfp.loc[dfp["atom"] == resonance]
    dfm_f = dfm.loc[dfm["Atom"] == resonance]
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    
    #print(D.head(5))
    # measure linear correction using all chemical shifts:
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m, c = np.polyfit(x, y, 1)
    print("All shifts:    m = %3.2f, c = %3.2f"%(m, c))
    
    # measure linear correction using apical loop shifts:
    D = D.loc[D["res"].isin(residue_nos)]
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m_, c_ = np.polyfit(x, y, 1)
    print("Loop shifts:    m = %3.2f, c = %3.2f"%(m_, c_))
    
    p_avg = np.mean(D.iloc[:, 4:4+20], axis=1)
    
    if resonance == "C6":
        fake_rows = [ [np.nan]*2 + [32] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [33] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [34] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [35] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [36] + ["C6"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    elif resonance == "C8":
        fake_rows = [ [np.nan]*2 + [29] + ["C8"] + [np.nan]*21, 
                      [np.nan]*2 + [30] + ["C8"] + [np.nan]*21,  
                      [np.nan]*2 + [31] + ["C8"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    # Apply linear correction to all predicted chemical shifts:
    if resonance == "C1'":
        #print(dfp_f.res)
        print(D.iloc[-2:-1, 4:4+20].to_numpy())
        print(" %f, %f"%(m,c))
        print(D.iloc[-2:-1, 4:4+20].to_numpy() * m + c)
        #print("X = ", x)
        #print("Y = ", y)
        #print(" %f, %f"%(m,c))
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    if resonance == "C1'":
        #print(dfp_f.res)
        print(All_pCS[-2:-1,:])
    #print(ES_indices)
    #print(GS_indices)
    GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices], axis=1))
    ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices], axis=1))
    GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices], axis=1))
    ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices], axis=1))
    #if resonance == "C1'":
        #print(GS_pCS_avg)
        #print(ES_pCS_avg)
    #if resonance == "C8":
    #    print(residue_nos)
    #    print("ES = ", ES_pCS_avg)
    #    print("GS = ", GS_pCS_avg)
    DW_res_no += residue_nos
    DW_atom += [resonance] * 8
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted


# merge and filter out residues to plot
df1 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

#print(df1)

DF = df1

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(width=2,length=6)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF.pdf")


d = DF.copy()
#d = d.loc[~((d.res== 35) & (d.atom == "C1'"))]
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
d["err"] = d.dw_avg - d.predicted_avg_dw
#print(d)
print()
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
print(d)
fig, ax = plt.subplots(figsize = (4, 2.5))
ax.errorbar(d.dw_avg, d.predicted_avg_dw, xerr=d.dw_err_avg, yerr=d.predict_std_dw, fmt='o', markersize=12,
           elinewidth=2, capthick=2, color="k", ecolor="k", capsize=6)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF_Correlation.pdf")

df_lowest_rmsd = d.copy()
print(df_lowest_rmsd)

d2 = df_lowest_rmsd.loc[~((df_lowest_rmsd.res== 35) & (df_lowest_rmsd.atom == "C1'"))]
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
rmsd = calc_rmsd(d2.dw_avg, d2.predicted_avg_dw)
print(np.corrcoef(d2.dw_avg, d2.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)

plt.show()
```

```python
# FARFAR-Unbiased wt-TAR RDC Optimized Ensemble (N=20) (Lowest RMSD Orca5 with clash filtering
# with VTSCF)
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to ensemble

dfp = pd.read_csv("data/Orca_DFT/wtTAR_Unbiased_NoClash_LowestRMSD_Ensemble_Filtered.csv").drop_duplicates()
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])

# Indices of the conformers (relative position in lib)
indices = np.array([332,  367,  636,  683,  686,  997, 1196, 2295, 2615, 2751, 3051, 
                    4151, 4348, 4460, 4793, 4801, 6962, 8322, 9002, 9615], dtype=int)
ES_indices = (indices >= 5000)
GS_indices = (indices < 5000)

# List of resonances & residues to plot for this analysis 
resonances = ["C1'", "C4'", "C5'", "C8", "C6"]
residue_nos = [29, 30, 31, 32, 33, 34, 35, 36]
residue_names = ["C", "C", "U", "G", "G", "G", "A", "G"]

DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

for resonance in resonances:
    dfp_f = dfp.loc[dfp["atom"] == resonance]
    dfm_f = dfm.loc[dfm["Atom"] == resonance]
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    
    #print(D.head(5))
    # measure linear correction using all chemical shifts:
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m, c = np.polyfit(x, y, 1)
    print("All shifts:    m = %3.2f, c = %3.2f"%(m, c))
    
    # measure linear correction using apical loop shifts:
    D = D.loc[D["res"].isin(residue_nos)]
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m_, c_ = np.polyfit(x, y, 1)
    print("Loop shifts:    m = %3.2f, c = %3.2f"%(m_, c_))
    
    
    p_avg = np.mean(D.iloc[:, 4:4+20], axis=1)
    #print(p_avg)
    #print(D["shift_avg"])
    
    if resonance == "C6":
        fake_rows = [ [np.nan]*2 + [32] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [33] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [34] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [35] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [36] + ["C6"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    elif resonance == "C8":
        fake_rows = [ [np.nan]*2 + [29] + ["C8"] + [np.nan]*21, 
                      [np.nan]*2 + [30] + ["C8"] + [np.nan]*21,  
                      [np.nan]*2 + [31] + ["C8"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    # Apply linear correction to all predicted chemical shifts:
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    #print(ES_indices)
    #print(GS_indices)
    GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices], axis=1))
    ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices], axis=1))
    GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices], axis=1))
    ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices], axis=1))
    #if resonance == "C8":
    #    print(residue_nos)
    #    print("ES = ", ES_pCS_avg)
    #    print("GS = ", GS_pCS_avg)
    DW_res_no += residue_nos
    DW_atom += [resonance] * 8
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted

# merge and filter out residues to plot
df1 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

#print(df1)

DF = df1

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(width=2,length=6)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF.pdf")


d = DF.copy()
#d = d.loc[~((d.res== 35) & (d.atom == "C1'"))]
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
d["err"] = d.dw_avg - d.predicted_avg_dw
#print(d)
print()
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
print(d)
fig, ax = plt.subplots(figsize = (5, 4))
ax.errorbar(d.dw_avg, d.predicted_avg_dw, xerr=d.dw_err_avg, yerr=d.predict_std_dw, fmt='o', markersize=14,
           elinewidth=2, capthick=2, color="k", ecolor="k", capsize=6)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF_Correlation.pdf")

df_lowest_rmsd = d.copy()
print(df_lowest_rmsd)

d2 = df_lowest_rmsd.loc[~((df_lowest_rmsd.res== 35) & (df_lowest_rmsd.atom == "C1'"))]
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
rmsd = calc_rmsd(d2.dw_avg, d2.predicted_avg_dw)
print(np.corrcoef(d2.dw_avg, d2.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)

```

```python
# FARFAR-Unbiased wt-TAR RDC Random ES/GS Ensemble (N=20) (Random Partition from Orca5)
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to randomly partitioned ensemble
dfp = pd.read_csv("data/Orca_DFT/wtTAR_GS_ES1G34syn_Unbiased_LowestRMSD_DFT_VTSCF.csv")
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])

# List of resonances & residues to plot for this analysis 
resonances = ["C1'", "C4'", "C5'", "C8", "C6"]
residue_nos = [29, 30, 31, 32, 33, 34, 35, 36]
residue_names = ["C", "C", "U", "G", "G", "G", "A", "G"]

N=100
import random
# Rewrite to forloop:
ES_indices_list = []
GS_indices_list = []
for i in range(N):
    np.random.seed(i*720)
    ES_indices_ = np.unique(random.sample(range(0, 20), 4))
    GS_indices_ = np.array([i for i in range(20) if i not in ES_indices_])
    ES_indices_list.append(ES_indices_)
    GS_indices_list.append(GS_indices_)



#print(ES_indices_list)
#print(GS_indices_list)
DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

for resonance in resonances:
    dfp_f = dfp.loc[dfp["atom"] == resonance]
    dfm_f = dfm.loc[dfm["Atom"] == resonance]
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    
    # measure linear correction using all chemical shifts:
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m, c = np.polyfit(x, y, 1)
    print("All shifts:    m = %3.2f, c = %3.2f"%(m, c))
    
    # measure linear correction using apical loop shifts:
    D = D.loc[D["res"].isin(residue_nos)]
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m_, c_ = np.polyfit(x, y, 1)
    print("Loop shifts:    m = %3.2f, c = %3.2f"%(m_, c_))
       
    p_avg = np.mean(D.iloc[:, 4:4+20], axis=1)
    
    if resonance == "C6":
        fake_rows = [ [np.nan]*2 + [32] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [33] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [34] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [35] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [36] + ["C6"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    elif resonance == "C8":
        fake_rows = [ [np.nan]*2 + [29] + ["C8"] + [np.nan]*21, 
                      [np.nan]*2 + [30] + ["C8"] + [np.nan]*21,  
                      [np.nan]*2 + [31] + ["C8"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    # Apply linear correction to all predicted chemical shifts:
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    
    # rewrite to for loop:
    
    
    for i in range(N):
        if i == 0:
            GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices_list[0]], axis=1))
            ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices_list[0]], axis=1))
            GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices_list[0]], axis=1))
            ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices_list[0]], axis=1))
        else:
            GS_pCS_avg += np.nan_to_num(np.nanmean(All_pCS[:, GS_indices_list[i]], axis=1))
            ES_pCS_avg += np.nan_to_num(np.nanmean(All_pCS[:, ES_indices_list[i]], axis=1))
            GS_pCS_std += np.nan_to_num(np.nanstd(All_pCS[:, GS_indices_list[i]], axis=1))
            ES_pCS_std += np.nan_to_num(np.nanstd(All_pCS[:, ES_indices_list[i]], axis=1))
    
    GS_pCS_avg = GS_pCS_avg/N
    ES_pCS_avg = ES_pCS_avg/N
    GS_pCS_std = GS_pCS_std/N
    ES_pCS_std = ES_pCS_std/N

    
    
    
    #if resonance == "C8":
    #    print(residue_nos)
    #    print("ES = ", ES_pCS_avg)
    #    print("GS = ", GS_pCS_avg)
    DW_res_no += residue_nos
    DW_atom += [resonance] * 8
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted


# merge and filter out residues to plot
rand_df3 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

DF = rand_df3

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(width=2,length=6)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
plt.savefig("plots/ensemble_vs_RD_dw_wtTARRandomPartitionEnsemble_Orca5.pdf")
#d["err"] = d.dw_avg - d.predicted_avg_dw
d = DF.copy()
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
d["err"] = d.dw_avg - d.predicted_avg_dw
print(d)

fig, ax = plt.subplots(figsize = (4, 2.5))
ax.errorbar(d.dw_avg, d.predicted_avg_dw, xerr=d.dw_err_avg, yerr=d.predict_std_dw, fmt='o', markersize=12,
           elinewidth=2, capthick=2, color="red", ecolor="k", capsize=6)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
plt.savefig("plots/ensemble_vs_RD_dw_wtTARRandomPartitionEnsemble_Orca5_Correlation.pdf")

```

```python
# FARFAR-Unbiased wt-TAR RDC Optimized Ensemble (N=20) Replicate after removal (Lowest RMSD Orca5 with VTSCF)
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to ensemble

dfp = pd.read_csv("data/Orca_DFT/wtTAR_Unbiased_LowestRMSD_Ensemble_Replicate.csv").drop_duplicates()
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])

# Indices of the conformers (relative position in lib)
indices = np.array([1]*14 + [5001]*6, dtype=int)
ES_indices = (indices >= 5000)
GS_indices = (indices < 5000)

# List of resonances & residues to plot for this analysis 
resonances = ["C1'", "C4'", "C5'", "C8", "C6"]
residue_nos = [29, 30, 31, 32, 33, 34, 35, 36]
residue_names = ["C", "C", "U", "G", "G", "G", "A", "G"]

DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

for resonance in resonances:
    dfp_f = dfp.loc[dfp["atom"] == resonance]
    dfm_f = dfm.loc[dfm["Atom"] == resonance]
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    
    #print(D.head(5))
    # measure linear correction using all chemical shifts:
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m, c = np.polyfit(x, y, 1)
    print("All shifts:    m = %3.2f, c = %3.2f"%(m, c))
    
    # measure linear correction using apical loop shifts:
    D = D.loc[D["res"].isin(residue_nos)]
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m_, c_ = np.polyfit(x, y, 1)
    print("Loop shifts:    m = %3.2f, c = %3.2f"%(m_, c_))
    
    
    p_avg = np.mean(D.iloc[:, 4:4+20], axis=1)
    #print(p_avg)
    #print(D["shift_avg"])
    
    if resonance == "C6":
        fake_rows = [ [np.nan]*2 + [32] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [33] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [34] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [35] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [36] + ["C6"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    elif resonance == "C8":
        fake_rows = [ [np.nan]*2 + [29] + ["C8"] + [np.nan]*21, 
                      [np.nan]*2 + [30] + ["C8"] + [np.nan]*21,  
                      [np.nan]*2 + [31] + ["C8"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    # Apply linear correction to all predicted chemical shifts:
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    #print(ES_indices)
    #print(GS_indices)
    GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices], axis=1))
    ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices], axis=1))
    GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices], axis=1))
    ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices], axis=1))
    #if resonance == "C8":
    #    print(residue_nos)
    #    print("ES = ", ES_pCS_avg)
    #    print("GS = ", GS_pCS_avg)
    DW_res_no += residue_nos
    DW_atom += [resonance] * 8
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted

# merge and filter out residues to plot
df1 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

#print(df1)

DF = df1

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(width=2,length=6)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF.pdf")


d = DF.copy()
#d = d.loc[~((d.res== 35) & (d.atom == "C1'"))]
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
d["err"] = d.dw_avg - d.predicted_avg_dw
#print(d)
print()
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
print(d)
fig, ax = plt.subplots(figsize = (5, 4))
ax.errorbar(d.dw_avg, d.predicted_avg_dw, xerr=d.dw_err_avg, yerr=d.predict_std_dw, fmt='o', markersize=14,
           elinewidth=2, capthick=2, color="k", ecolor="k", capsize=6)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF_Correlation.pdf")

df_lowest_rmsd = d.copy()
print(df_lowest_rmsd)

d2 = df_lowest_rmsd.loc[~((df_lowest_rmsd.res== 35) & (df_lowest_rmsd.atom == "C1'"))]
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
rmsd = calc_rmsd(d2.dw_avg, d2.predicted_avg_dw)
print(np.corrcoef(d2.dw_avg, d2.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)

```

```python
# FARFAR-Unbiased wt-TAR RDC Optimized Ensemble (N=20) (Lowest RMSD Orca5 with VTSCF)
# ALTERNATIVE LINEAR CORRECTION : PURINES / PYRIMIDINES
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to ensemble

dfp = pd.read_csv("data/Orca_DFT/wtTAR_GS_ES1G34syn_Unbiased_LowestRMSD_DFT_VTSCF.csv").drop_duplicates()
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])
print(dfd.shape)

# Indices of the conformers (relative position in lib)
indices = np.array([332,  367,  636,  683,  686,  997, 1196, 2295, 2615, 2751, 3051, 
                    4151, 4348, 4460, 4793, 4801, 6962, 8322, 9002, 9615], dtype=int)
ES_indices = (indices >= 5000)
GS_indices = (indices < 5000)

# List of resonances & residues to plot for this analysis 
resonances = dfd.atom.tolist()
residue_nos = dfd.res.tolist()
residue_names = ['C', 'U', 'U', 'U', 'G', 'G', 'A', 'A']

DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

LC = {
    "C-C1'":0.8,
    "C-C4'":1.35,
    "U-C1'":0.8,
    "U-C4'":1.35,
    "U-C6":0.64,
    "A-C1'":1.16,
    "A-C4'":0.68,
    "A-C8":1.05,
    "G-C1'":1.16,
    "G-C4'":0.68,
    "G-C8":1.05,
    
}

for resonance, residue_no, residue_name in zip(resonances, residue_nos, residue_names):
    dfp_f = dfp.loc[(dfp["atom"] == resonance) & (dfp["res"] == residue_no)]
    dfm_f = dfm.loc[(dfm["Atom"] == resonance) & (dfm["Resi"] == residue_no)]
    
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    #print(D)
    
    key = str(residue_name) + "-" + str(resonance)
    m = LC[key]
    c = 0
        
    # Apply linear correction to all predicted chemical shifts:
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    #print(ES_indices)
    #print(GS_indices)
    GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices], axis=1))
    ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices], axis=1))
    GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices], axis=1))
    ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices], axis=1))
    
    DW_res_no += [residue_no]
    DW_atom += [resonance]
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    #print(DW_avg_predicted)
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted

# merge and filter out residues to plot
df1 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

#print(df1)

DF = df1

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(width=2,length=6)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF.pdf")

d = DF.copy()
#d = d.loc[~((d.res== 35) & (d.atom == "C1'"))]
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
d["err"] = d.dw_avg - d.predicted_avg_dw
#print(d)
print()
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
print(d)
fig, ax = plt.subplots(figsize = (5, 4))
ax.errorbar(d.dw_avg, d.predicted_avg_dw, xerr=d.dw_err_avg, yerr=d.predict_std_dw, fmt='o', markersize=14,
           elinewidth=2, capthick=2, color="k", ecolor="k", capsize=6)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
#plt.savefig("plots/ensemble_vs_RD_dw_wtTARLowestRMSDEnsemble_Orca5_VTSCF_Correlation.pdf")

df_lowest_rmsd = d.copy()
print(df_lowest_rmsd)

d2 = df_lowest_rmsd.loc[~((df_lowest_rmsd.res== 35) & (df_lowest_rmsd.atom == "C1'"))]
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
rmsd = calc_rmsd(d2.dw_avg, d2.predicted_avg_dw)
print(np.corrcoef(d2.dw_avg, d2.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
plt.show()
```

```python
# FARFAR-Unbiased wt-TAR RDC Random ES/GS Ensemble (N=20) (Random Partition from Orca5)
# Read all Orca predicted chemical shifts from ensemble and measured values
# Apply linear correction to predicted values (LS Reg)
# Plot dw calculated from applying linear correction to randomly partitioned ensemble
dfp = pd.read_csv("data/Orca_DFT/wtTAR_GS_ES1G34syn_Unbiased_LowestRMSD_DFT_VTSCF_OutlierFiltered.csv")
dfp["shift_avg"] = np.nanmean(dfp.iloc[:, 2:], axis=1)
dfm = pd.read_csv("data/HIV_TAR_measured_cs.csv")
dfd = pd.read_csv("data/ES1_measured_dw.csv").dropna(subset=["dw_avg"])

# List of resonances & residues to plot for this analysis 
resonances = ["C1'", "C4'", "C5'", "C8", "C6"]
residue_nos = [29, 30, 31, 32, 33, 34, 35, 36]
residue_names = ["C", "C", "U", "G", "G", "G", "A", "G"]

N=10
import random
# Rewrite to forloop:
ES_indices_list = []
GS_indices_list = []
for i in range(N):
    np.random.seed(i*720)
    ES_indices_ = np.unique(random.sample(range(0, 20), 4))
    GS_indices_ = np.array([i for i in range(20) if i not in ES_indices_])
    ES_indices_list.append(ES_indices_)
    GS_indices_list.append(GS_indices_)



#print(ES_indices_list)
#print(GS_indices_list)
DW_res_no = []
DW_atom = []
DW_avg_predicted = []
DW_std_predicted = []

for resonance in resonances:
    dfp_f = dfp.loc[dfp["atom"] == resonance]
    dfm_f = dfm.loc[dfm["Atom"] == resonance]
    D = dfm_f.merge(dfp_f, left_on=["Resi", "Atom"], right_on=["res", "atom"], how="inner")
    D = D.rename(columns = {'Shift' : 'm_shift_avg', 'SDev': 'm_shift_dev'})
    D = D.drop(columns = ["Resn", "Resi", "Nuc", "Atom"])
    print("Resonance: %s"%(resonance))
    
    # measure linear correction using all chemical shifts:
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m, c = np.polyfit(x, y, 1)
    print("All shifts:    m = %3.2f, c = %3.2f"%(m, c))
    
    # measure linear correction using apical loop shifts:
    D = D.loc[D["res"].isin(residue_nos)]
    y = D["m_shift_avg"]
    x = D["shift_avg"]
    m_, c_ = np.polyfit(x, y, 1)
    print("Loop shifts:    m = %3.2f, c = %3.2f"%(m_, c_))
       
    p_avg = np.mean(D.iloc[:, 4:4+20], axis=1)
    
    if resonance == "C6":
        fake_rows = [ [np.nan]*2 + [32] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [33] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [34] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [35] + ["C6"] + [np.nan]*21, 
                      [np.nan]*2 + [36] + ["C6"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    elif resonance == "C8":
        fake_rows = [ [np.nan]*2 + [29] + ["C8"] + [np.nan]*21, 
                      [np.nan]*2 + [30] + ["C8"] + [np.nan]*21,  
                      [np.nan]*2 + [31] + ["C8"] + [np.nan]*21 ]
        D2 = pd.DataFrame(fake_rows, columns = D.columns)
        D = D.append(D2).sort_values(by=["res"])
        
    # Apply linear correction to all predicted chemical shifts:
    All_pCS = np.array(D.iloc[:, 4:4+20], dtype=float) * m + c
    
    # rewrite to for loop:
    
    
    for i in range(N):
        if i == 0:
            GS_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, GS_indices_list[0]], axis=1))
            ES_pCS_avg = np.nan_to_num(np.nanmean(All_pCS[:, ES_indices_list[0]], axis=1))
            GS_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, GS_indices_list[0]], axis=1))
            ES_pCS_std = np.nan_to_num(np.nanstd(All_pCS[:, ES_indices_list[0]], axis=1))
        else:
            GS_pCS_avg += np.nan_to_num(np.nanmean(All_pCS[:, GS_indices_list[i]], axis=1))
            ES_pCS_avg += np.nan_to_num(np.nanmean(All_pCS[:, ES_indices_list[i]], axis=1))
            GS_pCS_std += np.nan_to_num(np.nanstd(All_pCS[:, GS_indices_list[i]], axis=1))
            ES_pCS_std += np.nan_to_num(np.nanstd(All_pCS[:, ES_indices_list[i]], axis=1))
    
    GS_pCS_avg = GS_pCS_avg/N
    ES_pCS_avg = ES_pCS_avg/N
    GS_pCS_std = GS_pCS_std/N
    ES_pCS_std = ES_pCS_std/N

    
    
    
    #if resonance == "C8":
    #    print(residue_nos)
    #    print("ES = ", ES_pCS_avg)
    #    print("GS = ", GS_pCS_avg)
    DW_res_no += residue_nos
    DW_atom += [resonance] * 8
    DW_avg_predicted += (ES_pCS_avg - GS_pCS_avg).tolist()
    std_diff_means = np.sqrt((ES_pCS_std**2)/4 + (GS_pCS_std**2)/16).tolist()
    DW_std_predicted += std_diff_means

df = pd.DataFrame()
df["res"] = DW_res_no
df["atom"] = DW_atom
df["predicted_avg_dw"] = DW_avg_predicted
df["predict_std_dw"] = DW_std_predicted


# merge and filter out residues to plot
rand_df3 = df.merge(dfd, how="inner", on=["res", "atom"]).loc[:, ["res", "atom", "dw_avg", "dw_err_avg", 
                                    "predicted_avg_dw", "predict_std_dw"]].sort_values(by = ["res", "atom"])

DF = rand_df3

x_tick_pos = np.array(range(1, 120, 10)) [:8]
x_bar_pos = np.array(range(1, 120, 10))[:8]

x_tick_labs = ["C30-C1'", "U31-C1'", "U31-C4'", "U31-C6", "G34-C1'", "G34-C8", "A35-C1'", "A35-C4'"]

fig, ax = plt.subplots(figsize = (10, 4))

ax.bar(x = x_bar_pos-1.8, height=DF.dw_avg, width=3, color="black", yerr=DF.dw_err_avg,
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, edgecolor="black", label="Measured")
    
ax.bar(x = x_bar_pos+1.8, height=DF.predicted_avg_dw, width=3, color="black", fill=False, lw=2, yerr=DF.predict_std_dw, ecolor="black",
    capsize=6, error_kw = {'elinewidth':2, 'capthick':2}, label="Predicted")
ax.axhline(y=0, color="k", ls="--", lw=2)
ax.set_ylim([-7.5, 7.5])
ax.set_xticklabels(labels=x_tick_labs, fontsize=16)
ax.set_xticks(ticks=x_tick_pos)
ax.tick_params(width=2,length=6)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}$ (ppm)", fontsize=16)
plt.savefig("plots/ensemble_vs_RD_dw_wtTARRandomPartitionEnsemble_Orca5.pdf")
#d["err"] = d.dw_avg - d.predicted_avg_dw
d = DF.copy()
rmsd = calc_rmsd(d.dw_avg, d.predicted_avg_dw)
r2 = r2_score(d.dw_avg, d.predicted_avg_dw)
print(np.corrcoef(d.dw_avg, d.predicted_avg_dw)[0,1]**2)
print(r2)
print(rmsd)
d["err"] = d.dw_avg - d.predicted_avg_dw
print(d)

fig, ax = plt.subplots(figsize = (4, 3))
colors = ['red', 'red', 'blue', 'green', 'red', 'orange', 'red', 'blue']
msymbols = ["o", "o", "o", "s", "o", "s", "o", "o"]
for x,y,xerr,yerr,mc,ms in zip(d.dw_avg.tolist(), d.predicted_avg_dw.tolist(), d.dw_err_avg.tolist(), d.predict_std_dw.tolist(), colors, msymbols):
    ax.errorbar(x, y, xerr=xerr, yerr=yerr, fmt=ms, markersize=14, elinewidth=2, capthick=2, color=mc, mec='k', ecolor="k", capsize=6, linewidth=2)
ax.plot([-8, 8], [-8,8], color="k", ls="--", lw=2)
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_xticks([-3, 0, 3])
ax.set_yticks([-3, 0, 3])
ax.tick_params(width=2,length=6)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylabel(r"$\Delta{\rm \omega}_{\rm ens}$ (ppm)", fontsize=16)
ax.set_xlabel(r"$\Delta{\rm \omega}_{\rm RD}$ (ppm)", fontsize=16)
plt.savefig("plots/FINAL_ensemble_vs_RD_dw_wtTARRandomPartitionEnsemble_Orca5_Correlation.pdf")

d_random_partition = d.copy()
```
