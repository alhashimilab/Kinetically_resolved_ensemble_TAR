```python
# Defining required functions
def calc_rmsd(x, y):
    if len(x) == 0:
        return 0.0
    chi2 = sum((y - x) ** 2)
    rmsd = np.sqrt(chi2 / len(y))
    return rmsd

def calc_r2(x, y):
    if len(x) == 0:
        return 0.0
    correlation = r2_score(y, x)
    return correlation

def calc_r2_denovo(x, y):
    TSS = np.sum((x - np.mean(x))**2)
    RSS = np.sum((x - y)**2)
    return (1-RSS/TSS)

def diff(list1, list2):
    c = set(list1).union(set(list2))  # or c = set(list1) | set(list2)
    d = set(list1).intersection(set(list2))  # or d = set(list1) & set(list2)
    common = []
    for element in list1:
        if element in list2:
            common.append(element)
    return list(c - d), common
```

```python
# Read in json files:
df1 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate1-dssr.csv")
df2 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate2-dssr.csv")
df3 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate3-dssr.csv")

# Read in chemical shift files:
cs1 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate1.dft.csv")
cs2 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate2.dft.csv")
cs3 = pd.read_csv("data/ES1-Titrations-G34syn-Replicate3.dft.csv")
```

```python
# combine cs1 with df1:
shift_c1p = []
shift_aro = []
shift_c4p = []
for idx, row in df1.iterrows():
    resnum = row['nt_resnum']
    confnum = row['conf_no']
    back_idx = -1 - (20-confnum)
    
    c1p_ = cs1.loc[(cs1['res_no'] == resnum) & (cs1["atom"] == "C1'")].iloc[:,back_idx]
    c4p_ = cs1.loc[(cs1['res_no'] == resnum) & (cs1["atom"] == "C4'")].iloc[:,back_idx]
    aro_ = cs1.loc[(cs1['res_no'] == resnum) & ((cs1["atom"] == "C8")|(cs1["atom"] == "C6"))].iloc[:,back_idx]
    # C1'
    if c1p_.shape[0] != 1:
        shift_c1p += [np.nan]
    else:
        shift_c1p += c1p_.tolist()
    # C4'
    if c4p_.shape[0] != 1:
        shift_c4p += [np.nan]
    else:
        shift_c4p += c4p_.tolist()
    # Aromatic
    if aro_.shape[0] != 1:
        shift_aro += [np.nan]
    else:
        shift_aro += aro_.tolist()
shift_c1p = np.array(shift_c1p, dtype=float)
shift_c4p = np.array(shift_c4p, dtype=float)
shift_aro = np.array(shift_aro, dtype=float)
df1["shift_c1p"] = shift_c1p
df1["shift_c4p"] = shift_c4p
df1["shift_aro"] = shift_aro
```

```python
# Replicate 1:
# A35-C1': Plot just w for each conformer:

# use the linear correction previously observed:
m = 1.09
c = -14.21

# Defined ES and GS measured chemical shifts:
mAvg = 90.735
mESPop = 0.14
mGSPop = 1 - mESPop
mDw = 2.5

mES = (mAvg + mGSPop*mDw)/(mGSPop+mESPop)
mGS = mES - mDw
#print(mES, mGS)
df_t = df1.loc[(df1["nt_resnum"] == 35)]
fig, axs = plt.subplots(2,1, figsize=(10, 6))
ax=axs[0]
colors = ["blue"] * 16 + ["red"] * 4
ax.bar(x=df_t.conf_no, height=df_t.shift_c1p*m+c, color=colors, width=0.65)
#ax.bar(x=df_t.conf_no, height=df_t.shift_c1p*m+c, color=colors, width=0.65)
ax.bar(x=[21, 22], height=[mGS, mES], edgecolor=["blue", "red"], fill=False, width=0.65, lw=2)
ax.set_ylim([np.mean(df_t.shift_c1p*m +c)-6, np.mean(df_t.shift_c1p*m+c)+7])
ax.set_xlim([0.5, 22.5])
ax.tick_params(width=2, length=6)
ticks = ax.set_xticks(list(range(1, 23)))
ticklabels = ax.set_xticklabels(list(range(1,21)) + ["GS","ES"])
ax.axvline(x=16.5, color="k", ls="-", lw=2)
ax.axvline(x=20.5, color="k", ls="-", lw=2)
ax.axhline(np.nanmean(df_t.iloc[:16].shift_c1p*m+c), xmin=0, xmax=16.0/22, color="k", ls="--", lw=2)
ax.axhline(np.nanmean(df_t.iloc[-4:].shift_c1p*m+c), xmin=16.0/22, xmax=20.0/22, color="k", ls="--", lw=2)
ax.set_ylabel("CS (ppm)")

# A35: Plot Chi Angle:
#fig, ax = plt.subplots(figsize=(12, 4))
ax=axs[1]
colors = ["blue"] * 16 + ["red"] * 4
ax.bar(x=df_t.conf_no, height=df_t["phase_angle"], color=colors, width=0.65)
ax.set_xlim([0.5, 22.5])
ax.tick_params(width=2, length=6)
ticks = ax.set_xticks(list(range(1, 23)))
ax.axvline(x=16.5, color="k", ls="-", lw=2)
ax.axvline(x=20.5, color="k", ls="-", lw=2)
ax.set_ylabel("Phase Angle")

#print(df_t.shape)
#print(df_t.loc[df_t.sugar_class == "~C2'-endo"].shape)
#print(df_t.loc[df_t.sugar_class == "~C2'-endo"].loc[df_t.conf_no.isin(list(range(1,17)))].shape)

#sys.exit(0)
fig, ax = plt.subplots(figsize=(4, 2.5))
GS_c2p = np.mean(df_t.loc[df_t.sugar_class == "~C2'-endo"].loc[df_t.conf_no.isin(list(range(1,17)))].shift_c1p*m+c)
GS_c3p = np.mean(df_t.loc[df_t.sugar_class == "~C3'-endo"].loc[df_t.conf_no.isin(list(range(1,17)))].shift_c1p*m+c)
ES_c2p = np.mean(df_t.loc[df_t.sugar_class == "~C2'-endo"].loc[df_t.conf_no.isin(list(range(17,21)))].shift_c1p*m+c)
ES_c3p = np.mean(df_t.loc[df_t.sugar_class == "~C3'-endo"].loc[df_t.conf_no.isin(list(range(17,21)))].shift_c1p*m+c)
ax.bar(x=[1,2,3,4, 5, 6], height=[GS_c2p, GS_c3p,mGS, ES_c2p, ES_c3p, mES], width=0.5, color=["blue"]*2+["k"] + ["red"]*2+["k"])
ax.set_ylim([89, 93])
ax.tick_params(width=2, length=6)
ticks = ax.set_xticks(list(range(1, 7)))
```
