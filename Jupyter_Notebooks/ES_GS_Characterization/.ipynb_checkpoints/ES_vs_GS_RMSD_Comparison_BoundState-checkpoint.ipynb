```python
# Plot ABG 2D Histograms for FARFAR-Unbiased Ensemble (ES only)
fig, ax = pl.subplots(1,figsize=(8,4))
binwidth = 20

rmsd = pd.read_csv("data/wtTAR_Ensemble_SEC_RMSD.csv", sep=",")
rmsd_ES = rmsd.loc[rmsd["State"] == "ES"]
rmsd_GS = rmsd.loc[rmsd["State"] == "GS"]


sns.distplot(a=rmsd_ES["Backbone_RMSD"], kde=True, ax=ax, bins=10, color="red")
sns.distplot(a=rmsd_GS["Backbone_RMSD"], kde=True, ax=ax, bins=10, color="blue")
#ax.hist(x=rmsd_GS["Backbone_RMSD"], bins=20, color="blue", alpha=0.5)

#ax.hist(x=A_GS, range=[-180, 180], bins=25, color="blue", alpha=0.5, density=True)
ax.set_xlabel("Backbone RMSD with Bound State (A)")
ax.set_xlim([1.5, 5])
#ax.set_ylim(0,10)


ax.tick_params(width=2,length=6)
plt.savefig("plots/Bound_state_RMSD_ES_vs_GS_Hist1D.pdf")
```

```python
df = pd.read_table("data/wtTAR_RMSD_BoundState.cpptrajout", delim_whitespace=True)
df2 = pd.read_csv("data/index_map.csv")
df["sas_idx"] = df2.sas_index
GS_mask = df["sas_idx"] < 5000
ES_mask = df["sas_idx"] >= 5000
print(df.loc[ES_mask].head)
```

```python
#fig, ax = plt.subplots()
sns.histplot(data=df.loc[GS_mask], x="all_heavy_atom", stat="density", binwidth=0.2, binrange=[3.01, 7.99], kde=False, color="green", alpha=0.35)
sns.histplot(data=df.loc[ES_mask], x="all_heavy_atom", stat="desnity", binwidth=0.2, binrange=[3.01, 7.99], kde=False, color="magenta", alpha=0.35)
plt.xlim([2.9, 7.9])
#plt.ylim([3.1, 7.9])

plt.show()
```

```python
fig, ax = plt.subplots(1,figsize=(2,4))
vp = ax.violinplot([df.loc[GS_mask].all_heavy_atom, df.loc[ES_mask].all_heavy_atom], positions = [1, 2], widths=0.75,
              showmeans=True)
colors = ['green', 'magenta']
counter = 0

vp['cmeans'].set_color('black')
vp['cmins'].set_color('black')
vp['cmaxes'].set_color('black')
vp['cbars'].set_color('black')


for pc in vp['bodies']:
    pc.set_facecolor(colors[counter])
    pc.set_alpha(0.6)
    counter += 1
```

```python
fig, ax = plt.subplots(1,figsize=(2,4))
vp = ax.violinplot([df.loc[GS_mask].full_backbone_heavy_atom, df.loc[ES_mask].full_backbone_heavy_atom], positions = [1, 2], widths=0.75,
              showmeans=True)
colors = ['green', 'magenta']
counter = 0

vp['cmeans'].set_color('black')
vp['cmins'].set_color('black')
vp['cmaxes'].set_color('black')
vp['cbars'].set_color('black')


for pc in vp['bodies']:
    pc.set_facecolor(colors[counter])
    pc.set_alpha(0.6)
    counter += 1
```
