```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib as mpl
```

```python
df = pd.read_excel("data/Ensemble_character.xlsx", sheet_name="Liz_S2").iloc[:6, :9]

fig, ax = plt.subplots(figsize=(12,5))
x= np.array([1, 6, 11, 16, 21, 26])
ax.bar(x, df["S2rel_meas_liz"], width=0.75, color="k", label="Measured")
ax.bar(x+1, df["S2_rel_pred_ENS"], width=0.75, color="black", label="Ensemble", fill=False, lw=2)
ax.bar(x+2, df["S2_rel_pred_GS"], width=0.75, color="green", label="GS", fill=False, hatch="//", edgecolor='green', lw=2)
ax.bar(x+3, df["S2_rel_pred_ES"], width=0.75, color="magenta", label="ES", fill=False, hatch="//", edgecolor='magenta', lw=2)
ax.legend(bbox_to_anchor=(1.5, 1.15))
ax.set_ylim([0,1.1])
ax.set_xlim([0, 30.1])

ax.set_xticks((x+1.5).tolist(), labels=["C30-C6", "G32-C8", "G33-C8", "G34-C8", "A35-C8", "A35-C2"] )
ax.set_ylabel(r"$ S^{2}_{rel}$")
ax.tick_params(width=2,length=6)
#plt.savefig("S2_rel_mvp.pdf")
plt.show()
```
