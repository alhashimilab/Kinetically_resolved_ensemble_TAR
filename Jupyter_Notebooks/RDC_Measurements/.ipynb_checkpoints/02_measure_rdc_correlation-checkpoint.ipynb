```python
# import required libraries
import os, sys
import numpy as np, pandas as pd, seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import pearsonr
```

```python
# Load All measured RDC Data
E0_rdc = pd.read_csv("data/E0-wt-TAR.sas.csv")
EI22_rdc = pd.read_csv("data/EI22-wt-TAR.sas.csv")
EII3_rdc = pd.read_csv("data/EII3-wt-TAR.sas.csv")
EII13_rdc = pd.read_csv("data/EII13-wt-TAR.sas.csv")
```

```python
# Define function for scaling two sets of RDCs :
def get_scaled_rdcs(m1, m2):
    L = np.dot(m1, m2)/np.dot(m2, m2)
    scaled_m2 = m2 * L    
    return m1.copy(), scaled_m2.copy()
```

```python
# Settings for matplotlib
plt.rcParams["font.family"] = "Arial"
mpl.rcParams['axes.linewidth'] = 2
plt.rcParams.update({'font.size': 20})
```

```python
# Plot E0 vs EI22 All Motifs:
df = E0_rdc.merge(EI22_rdc, on=["res_id_1", "res_name_1", "atom_1", "res_id_2", "res_name_2", "atom_2", "bond_vector"], 
                  suffixes=["_x", "_y"], how="inner")
maskH1 = df.res_id_1.isin([18, 19, 20, 21, 22, 40, 41, 42, 43, 44])
maskH2 = df.res_id_1.isin([26, 27, 28, 29, 36, 37, 38, 39])
maskB = df.res_id_1.isin([23, 24, 25])
maskL = df.res_id_1.isin([30, 31, 32, 33, 34, 35])

x, y = get_scaled_rdcs(df.rdc_x, df.rdc_y)
df["scaled_rdc_x"] = x
df["scaled_rdc_y"] = y

# Plot E0 measured vs back-calculated from RAMAH:
fig, ax = plt.subplots(figsize=(6,4))

# H1:
x = df.loc[maskH1].scaled_rdc_x
y = df.loc[maskH1].scaled_rdc_y
rho = pearsonr(x, y)[0]
sns.scatterplot(x=y, y=x, color="red", marker="o", s=120, 
                label=r"$\rho_{H1}$ = %3.2f"%(rho), edgecolor=None)

# H2:
x = df.loc[maskH2].scaled_rdc_x
y = df.loc[maskH2].scaled_rdc_y
rho = pearsonr(x, y)[0]
sns.scatterplot(x=y, y=x, color="blue", marker="o", s=120, 
                label=r"$\rho_{H2}$ = %3.2f"%(rho), edgecolor=None)

# B:
x = df.loc[maskB].scaled_rdc_x
y = df.loc[maskB].scaled_rdc_y
rho = pearsonr(x, y)[0]
sns.scatterplot(x=y, y=x, color="orange", marker="o", s=120, 
                label=r"$\rho_B$ = %3.2f"%(rho), edgecolor=None)

# L:
x = df.loc[maskL].scaled_rdc_x
y = df.loc[maskL].scaled_rdc_y
rho = pearsonr(x, y)[0]
sns.scatterplot(x=y, y=x, color="green", marker="o", s=120, 
                label=r"$\rho_L$ = %3.2f"%(rho), edgecolor=None)


ax.tick_params(width=2,length=6)
ax.set_ylabel("E0 RDC (Hz)")
ax.set_xlabel("EI22 RDC (Hz)")
ax.set_xlim([-45, 45])
ax.set_ylim([-45, 45])
plt.plot([-45, 45], [-45, 45], "k--", lw=2)
plt.legend(loc="upper left", prop={'size': 16})

plt.savefig("plots/E0-vs-EI22-RDC-Corr.pdf")
```
